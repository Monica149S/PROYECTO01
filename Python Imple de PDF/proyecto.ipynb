## Generando archivos PDF del presupuesto
* Crear el archivo PDF
* Definir el tipo de letra
* Llenar el documento
* Cargar el entregable

```python
!pip install fpdf
```

```python
from fpdf import FPDF
```

```python
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial")
pdf.image("TemplateImp.png",0, 0, 210, 300)
pdf.text(115, 166, proyecto)
pdf.text(115, 184, horas_estimadas)
pdf.text(115, 200, valor_hora)
pdf.text(115, 215, plazo_estimado)
pdf.text(115, 230,str(valor_total))
pdf.output("Presupuesto.pdf")
```
